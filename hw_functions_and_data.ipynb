### Задание 1 

```python
import pandas as pd 
```

```python
rating = pd.read_csv('/Users/aleksandrsvistunov/Downloads/ml-latest-small/ratings.csv', sep = ',')
rating
```

```python
def classification(data):
    if data <= 2:
        return "Низкий рейтинг"
    elif data <= 4:
        return "Средний рейтинг"
    else:
        return "Высокий рейтинг"
```

### Задание 2

```python
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
keywords = pd.read_csv("/Users/aleksandrsvistunov/Downloads/Python_11_functions_groupby/keywords.csv")
keywords.tail(5)
```

```python
row1 = pd.DataFrame( { 'keyword' : ['погода в тула на неделю' ], 'shows' : [3688] })
row1
```

```python
def reg_class(data):
    if any( city in data for city in geo_data["Центр"]):
        return 'Центр'
    elif any( city in data for city in geo_data["Северо-Запад"]):
        return 'Северо-Запад'
    elif any( city in data for city in geo_data["Дальний Восток"]):
        return 'Дальний Восток'
    else:
        return 'undefined'
```

<a id='final_2'></a>

```python
keywords['region'] = keywords.keyword.apply(reg_class)
keywords[keywords['region'] != 'undefined' ]
```

### Задание 3

```python
movies = pd.read_csv('/Users/aleksandrsvistunov/Downloads/ml-latest-small/movies.csv', sep = ',')
movies
```

```python
def production_year(data):
    years = list(range(1950,2011))
    for year in years:
        if str(year) in data:
            return year
    return 1900
```

```python
movies['year'] = movies.title.apply(production_year)
movies
```

```python
rating_year = pd.merge( movies, rating, how = 'inner' )
```

<a id='final_3'></a>

```python
(rating_year[['year','rating']]
     .groupby('year')
     .mean()
     .sort_values( by = 'rating',ascending = False))
```
